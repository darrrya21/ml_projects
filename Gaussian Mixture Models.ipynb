```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
X, y = make_blobs(n_samples=400, centers=4, cluster_std=0.6, random_state=0)
```

```python
from sklearn.preprocessing import scale
```

```python
X = X[:, ::-1]
```

```python
plt.figure(figsize=(12, 7))
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X)
labels = kmeans.predict(X)
```

```python
labels
```

```python
y
```

```python
mapping = {1: 3, 0: 2, 3: 0, 2: 1}
```

```python
labels_mapped = [mapping[val] for val in labels]
```

```python
from sklearn.metrics import accuracy_score
```

```python
(y == labels_mapped).all()
```

```python
accuracy_score(y, labels_mapped)
```

```python
from scipy.spatial.distance import cdist
from matplotlib.patches import Circle
```

```python
radius = [cdist(X[labels == i], [center]).max() for i, center in enumerate(kmeans.cluster_centers_)]
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:, 0], X[:, 1], c=labels, zorder=2)
axes.set_xlim(-5, )
for center, rad in zip(kmeans.cluster_centers_, radius):
    axes.add_patch(plt.Circle(center, rad, alpha=0.3, fc='gray', zorder=1, edgecolor='black'))
```

```python
from sklearn.mixture import GaussianMixture
```

```python
gmm = GaussianMixture(n_components=4)
gmm.fit(X)
```

```python
labels_gmm = gmm.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels_gmm)
```
